```python
import numpy as np
import matplotlib.pylab as plt

import os

from yolo import YOLO
from PIL import Image
import time
from IPython.display import clear_output
```

```python
ext = '_mouse'
yoloObj = YOLO(model_path = "logs/trained_weights_stage_0" + ext + ".h5",
     classes_path = 'model_data/' +ext[1:]+ '_classes.txt' ,
     anchors_path = 'model_data/yolo_anchors' + ext + '.txt',
    model_image_size = (64,192),
              gpu_num = 0,
              score = 0.6,
              iou = 0.8)
```

```python
image_path = "/Users/raminanushiravani/Desktop/mouse_videos/healthy/healthy_NSG336_1_20cms_0degUP/frames/"
files = sorted(os.listdir(image_path), key = lambda x : int(x.split('.')[0]))
files_abs_path = [image_path+f for f in files ]
```

```python
for f in files_abs_path:
    figure, ax = plt.subplots(1,figsize=(40,20))
    image = Image.open(f)
    image_data = np.array(image)
    image_ = Image.fromarray(image_data.astype('uint8'), 'RGB')
    i = yoloObj.detect_image(image_)
    ax.imshow(i)
    plt.show()
    time.sleep(0.01)
    clear_output(wait=True)
```
